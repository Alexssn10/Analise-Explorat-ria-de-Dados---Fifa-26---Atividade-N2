## 3. Limpeza e Pr√©-processamento
### 3.1 Valores ausentes

```python
df.isna().sum().sort_values(ascending=False).head(20)
```

```python
# Adicionando informa√ß√µes para jogadores n√£o convocados
df["nation_team_id"] = df["nation_team_id"].fillna("Sem Sele√ß√£o")
df["nation_position"] = df["nation_position"].fillna("Sem Sele√ß√£o")
df["nation_jersey_number"] = df["nation_jersey_number"].fillna("Sem Sele√ß√£o")

# Adicionando informa√ß√µes para jogadores que n√£o foram emprestados
df["club_loaned_from"] = df["club_loaned_from"].fillna("N√£o emprestado")
# Adicionando o valor de 0 para jogadores de linha nos atributos de goleiros
df["goalkeeping_speed"] = df["goalkeeping_speed"].fillna(0)
# Adicionando a informa√ß√£o para jogadores que n√£o possuem tra√ßos que os definem, "jogadores b√°sicos"
df["player_traits"] = df["player_traits"].fillna("Nenhum")

# Preenchendo os jogadores que n√£o tem posi√ß√£o e que possuem alguns dados faltando com a m√©dia
for col in ["pace","physic","defending","dribbling","passing","shooting"]:
    df[col] = df.groupby("player_positions")[col].transform(lambda x: x.fillna(x.mean()))
# Jogadores com aus√™ncia de cla√∫sula n√£o ¬¥um problema, mas os que possuem podem ter utilidade para previs√µes futuras
df["release_clause_eur"] = df["release_clause_eur"].fillna(0)
# Mesmo caso da aus√™ncia de cla√∫sula,  pode ser usado para previs√µes futuras
df["club_joined_date"] = df["club_joined_date"].fillna("Desconhecido")
# Mostra jogadores livres no mercado
for col in ["club_name","club_contract_valid_until_year","club_jersey_number","club_position"]:
    df[col] = df[col].fillna("Sem clube")

# Remover colunas realmente in√∫teis
df = df.drop(columns=["work_rate", "player_tags"])
```

```python
num_cols = df_limpo.select_dtypes(include=["number"]).columns
cat_cols = df_limpo.select_dtypes(include=["object", "category"]).columns

# 2) Estat√≠sticas descritivas para NUM√âRICAS 
stats_num = pd.DataFrame({
    "count": df_limpo[num_cols].count(),
    "mean": df_limpo[num_cols].mean(),
    "median": df_limpo[num_cols].median(),
    "std": df_limpo[num_cols].std(),
    "min": df_limpo[num_cols].min(),
    "q1": df_limpo[num_cols].quantile(0.25),
    "q3": df_limpo[num_cols].quantile(0.75),
    "max": df_limpo[num_cols].max(),
})
stats_num["iqr"] = stats_num["q3"] - stats_num["q1"]  # intervalo interquartil
stats_num = stats_num.round(3)

print("\n=== Estat√≠sticas num√©ricas ===")
print(stats_num)

# 3) Frequ√™ncias para CATEG√ìRICAS 

freq_list = []
N = len(df_limpo)

# Selecionar colunas categ√≥ricas
cat_cols = df_limpo.select_dtypes(include=['object', 'category']).columns

for c in cat_cols:
    vc = df_limpo[c].value_counts(dropna=False)

    # Cria as colunas 'categoria' e 'freq' de forma est√°vel
    freq = vc.rename_axis("categoria").reset_index(name="freq")

    # Trata valores NaN como texto
    freq["categoria"] = freq["categoria"].astype(object).where(freq["categoria"].notna(), "NaN")

    # Adiciona informa√ß√µes da coluna e percentual
    freq["coluna"] = c
    freq["percent"] = (freq["freq"] / N * 100).round(2)

    # Adiciona ao acumulador
    freq_list.append(freq[["coluna", "categoria", "freq", "percent"]])

# Concatenar tudo em um √∫nico DataFrame final
freq_cat = pd.concat(freq_list, ignore_index=True)

# exemplo das 5 primeiras colunas
print("\n=== Frequ√™ncias categ√≥ricas (exemplo de at√© 5 colunas) ===")
for c in cat_cols[:5]:
    print(f"\n-- {c} --")
    print(freq_cat[freq_cat["coluna"] == c].head(10))

```

```python
for col in df_limpo.select_dtypes(include=['number']).columns:
    plt.figure(figsize=(8, 5))
    sns.histplot(df_limpo[col], kde=True, bins=30)
    plt.title(f'Distribui√ß√£o de {col}')
    plt.xlabel(col)
    plt.ylabel('Frequ√™ncia')
    plt.show()

# ============ GR√ÅFICOS DE BARRAS (vari√°veis categ√≥ricas) ============
for col in df_limpo.select_dtypes(include=['object', 'category']).columns:
    plt.figure(figsize=(10, 5))
    top_values = df_limpo[col].value_counts().head(10)  # top 10 categorias
    sns.barplot(x=top_values.values, y=top_values.index, palette='viridis')
    plt.title(f'Frequ√™ncia das principais categorias em {col}')
    plt.xlabel('Contagem')
    plt.ylabel(col)
    plt.show()

# ============ SCATTERPLOTS (rela√ß√µes entre vari√°veis num√©ricas) ============
num_cols = df_limpo.select_dtypes(include=['number']).columns

#valor de mercado x overall
if {'value_eur', 'overall'}.issubset(num_cols):
    plt.figure(figsize=(8, 6))
    sns.scatterplot(data=df_limpo, x='overall', y='value_eur', alpha=0.6)
    plt.title('Rela√ß√£o entre Overall e Valor de Mercado (‚Ç¨)')
    plt.xlabel('Overall')
    plt.ylabel('Valor (‚Ç¨)')
    plt.show()

# idade x overall
if {'age', 'overall'}.issubset(num_cols):
    plt.figure(figsize=(8, 6))
    sns.scatterplot(data=df_limpo, x='age', y='overall', alpha=0.6)
    plt.title('Rela√ß√£o entre Idade e Overall')
    plt.xlabel('Idade')
    plt.ylabel('Overall')
    plt.show()
```
